```python
#функция для отрисоки гистограмм и boxplot для количественных переменных
def plot_hist(data, col_column):
    
    rows = len(col_column)
    f, ax = plt.subplots(rows, 2, figsize=(10, 15))
    f.tight_layout()
    f.set_figheight(80)
    f.set_figwidth(14)
    plt.rcParams.update({'font.size': 18})
    
    for i, col in enumerate(col_column):         
        sns.histplot(data[col], kde=True, bins=24, ax = ax[i, 0])                    
        sns.boxplot(data[col], ax = ax[i, 1])

        ax[i, 0].set_xlabel(col)
        ax[i, 1].set_xlabel(col)
        ax[i, 0].set_ylabel('Количество')
    plt.suptitle("Гистограмма и ящик с усами для количественных данных", fontsize=22, y=1.01)
    plt.show()
```

### Проверка формулы вычисления эффективности обогащения

Рассчитаем эффективность обогащения по формуле:

```python
#введем переменные для расчета
C = train['rougher.output.concentrate_au']
F = train['rougher.input.feed_au']
T = train['rougher.output.tail_au']
```

```python
rougher_output_recovery_calc = (C*(F-T))/(F*(C-T))*100
rougher_output_recovery_calc
```

```python
MAE = mean_absolute_error(train['rougher.output.recovery'], rougher_output_recovery_calc)
MAE
```

**Предварительный вывод:**  
Значение ошибки очень мало, следовательно можем считать формулу вычисления эффективности верной.

### Cуммарная концентрация всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах. 

```python
#посчитаем суммарную концентрацию на каждом этапе

train_new['input.concentrate_sum'] = (train_new['rougher.input.feed_sol'] 
                                      + train_new['rougher.input.feed_ag'] 
                                      + train_new['rougher.input.feed_au'] 
                                      + train_new['rougher.input.feed_pb'])

train_new['rougher.output.concentrate_sum'] = (train_new['rougher.output.concentrate_sol'] 
                                               + train_new['rougher.output.concentrate_ag'] 
                                               + train_new['rougher.output.concentrate_au']
                                               + train_new['rougher.output.concentrate_pb'])

train_new['final.output.concentrate_sum'] = (train_new['final.output.concentrate_sol'] 
                                               + train_new['final.output.concentrate_ag'] 
                                               + train_new['final.output.concentrate_au']
                                               + train_new['final.output.concentrate_pb'])

#проверим
train_new
```

```python
fig, axes = plt.subplots(1, 3,  figsize=(16,7))
fig.suptitle('Суммарная концентрации элементов на разных стадиях')

sns.boxplot(data=train_new['input.concentrate_sum'], ax=axes[0], color='g').set(title='Исходное сырье', 
                                                                     ylabel='Концентрация, %');

sns.boxplot(data=train_new['rougher.output.concentrate_sum'], ax=axes[1], color='g').set(title='Черновой концентрат', 
                                                                              ylabel='Концентрация, %');

sns.boxplot(data=train_new['final.output.concentrate_sum'], ax=axes[2], color='g').set(title='Финальный концентрат', 
                                                                            ylabel='Концентрация, %');
```

```python
plt.figure(figsize=(16,7))
sns.histplot(train_new['input.concentrate_sum'], kde=True, label = 'исходное сырье', color='y')
sns.histplot(train_new['rougher.output.concentrate_sum'], kde=True, label = 'черновой концентрат', color='g')
sns.histplot(train_new['final.output.concentrate_sum'], kde=True, label = 'финальный концентрат', color='b')

plt.title('Распределение суммарной концентрации на разных стация процесса')
plt.ylabel('Количество')
plt.xlabel('Суммарная концентрация, %')
plt.legend()
plt.show()
```

**Предварительный вывод:**  
Видим, что на всех стадиях есть всплеск на нулевом значении. Эти данные следует удалить.

## Модель

### Функция для вычисления итогового sMAPE

```python
def sMAPE(y, y_pred):
    return (1/len(y))*sum( (abs(y - y_pred)) / ((abs(y)+abs(y_pred))/2) )*100   
```

```python
def final_sMAPE(sMAPE_crude, sMAPE_final):
    return 0.25*sMAPE_crude + 0.75*sMAPE_final
```

```python
def sMAPE_scoring(estimator, X, y):
    y_pred = estimator.predict(X)
    return sMAPE(y, y_pred)
```
